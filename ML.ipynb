```python
Question 1: Explain the differences between AI, ML, Deep Learning (DL), and DataScience (DS).

ANS- 1. Artificial Intelligence (AI)

Definition:
AI is the broad field of creating machines or software that can perform tasks that typically require human intelligence — such as reasoning, learning, decision-making, problem-solving, and language understanding.

Key Points:

Scope: The largest umbrella term; includes ML, DL, robotics, expert systems, etc.

Goal: Mimic or simulate human-like intelligence.

Examples: Chatbots, autonomous cars, speech recognition, chess-playing programs.

2. Machine Learning (ML)

Definition:
A subset of AI where systems learn from data and improve performance over time without being explicitly programmed for every rule.

Key Points:

Scope: AI technique that focuses on pattern detection and prediction.

Goal: Use algorithms to learn from past data to make predictions or decisions.

Examples: Spam email filtering, recommendation systems, fraud detection.

3. Deep Learning (DL)

Definition:
A subset of ML that uses artificial neural networks with multiple layers (deep architectures) to automatically learn complex patterns from large datasets.

Key Points:

Scope: Special kind of ML — great for handling unstructured data like images, audio, text.

Goal: Learn hierarchical feature representations directly from raw data.

Examples: Facial recognition, voice assistants, language translation, image captioning.

4. Data Science (DS)

Definition:
An interdisciplinary field focused on extracting insights and knowledge from data using methods from statistics, ML, AI, and domain expertise.

Key Points:

Scope: Broader than AI/ML — includes data cleaning, visualization, statistical analysis, and modeling.

Goal: Turn raw data into actionable insights for decision-making.

Examples: Sales forecasting, customer segmentation, business analytics dashboards.


Question 2: What are the types of machine learning? Describe each with one
real-world example.

Ans-  1. Supervised Learning

Definition:
The algorithm learns from labeled data — meaning both input and correct output are provided during training.
It tries to map inputs → outputs so it can predict for new data.

Key Points:

Learns with guidance (“supervised”).

Used for prediction and classification.

Involves two main tasks:

Regression → Predict continuous values (e.g., price, temperature).

Classification → Predict discrete labels (e.g., spam/not spam).

Example:

Email spam detection – The model is trained with examples of “spam” and “not spam” emails, then predicts new emails’ categories.

2. Unsupervised Learning

Definition:
The algorithm learns from unlabeled data — only the inputs are given, no correct output.
It tries to find hidden patterns or structure in data.

Key Points:

No explicit guidance.

Common tasks:

Clustering → Grouping similar data points.

Dimensionality Reduction → Simplifying data while keeping important features.

Example:

Customer segmentation – An e-commerce company groups customers based on their purchasing behavior to target marketing campaigns.

3. Semi-Supervised Learning

Definition:
Uses a small amount of labeled data plus a large amount of unlabeled data for training.
It combines the strengths of supervised and unsupervised learning.

Key Points:

Useful when labeling data is expensive or time-consuming.

Improves accuracy when fully labeled datasets are scarce.

Example:

Medical diagnosis – A few labeled MRI scans (with confirmed diagnoses) plus many unlabeled scans help train a model for disease detection.

4. Reinforcement Learning (RL)

Definition:
The model (agent) learns by interacting with an environment and receiving rewards or penalties based on its actions.
It aims to maximize cumulative reward over time.

Key Points:

No fixed dataset — learns from trial and error.

Focused on decision-making and sequences of actions.

Example:

Self-driving cars – The AI learns to drive by getting rewards for safe driving and penalties for collisions.

Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine
learning.

Ans-  1. Overfitting

Definition:
When a model learns too much detail from the training data — including noise and random fluctuations — so it performs very well on training data but poorly on new (test) data.

Key Points:

Model is too complex (too many parameters).

Memorizes the training set instead of generalizing.

Low training error, high test error.

Example:
A decision tree that splits until each leaf has only one training sample — perfect accuracy in training, but fails badly on unseen data.

2. Underfitting

Definition:
When a model is too simple to capture the underlying patterns in the data, so it performs poorly on both training and test data.

Key Points:

Model is not complex enough (too few parameters).

Fails to learn important relationships in the data.

High training error, high test error.

Example:
Trying to fit a straight line to data that has a curved relationship.

3. Bias–Variance Tradeoff

Definition:
A fundamental concept describing how model errors are influenced by two factors:

Bias → Error from incorrect assumptions in the learning algorithm (too simple a model → underfitting).

Variance → Error from too much sensitivity to small variations in the training data (too complex a model → overfitting).

Tradeoff:

Increasing model complexity reduces bias but increases variance.

Decreasing model complexity reduces variance but increases bias.

The goal is to find the sweet spot where both bias and variance are balanced for best generalization.

Question 4: What are outliers in a dataset, and list three common techniques for handling them.

Ans-  Definition:
Outliers are data points that differ significantly from most of the other observations in a dataset.
They can occur due to:

Measurement errors (e.g., faulty sensors, typos)

Natural variation (e.g., rare but valid cases)

Data entry errors

Why They Matter:

Can skew statistical measures (mean, variance).

Can mislead model training and reduce accuracy.

Sometimes are important insights (e.g., fraud detection).

Three Common Techniques to Handle Outliers
1. Removal of Outliers

Remove extreme values if they are due to errors or irrelevant to analysis.

Method: Use statistical thresholds like Z-score (> 3 or < -3) or IQR method.

Example: Removing height values over 3 meters in human height dataset.

2. Transformation

Apply mathematical transformations to reduce the impact of extreme values.

Methods: Logarithmic, Square Root, or Box-Cox transformations.

Example: Applying log transformation to income data to handle skewness.

3. Capping / Winsorization

Replace extreme values with a maximum/minimum threshold.

Method: Set values above the 95th percentile to the 95th percentile value, and below the 5th percentile to the 5th percentile value.

Example: In salary data, cap all salaries above ₹50L at ₹50L.

Question 5: Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data.

Ans-  Handling Missing Values — Process

Identify Missing Data

Check for missing values using tools like isnull() in Pandas or .describe() for summary statistics.

Determine the Cause

Missing completely at random (MCAR) → No relationship to other data.

Missing at random (MAR) → Related to some observed variables.

Missing not at random (MNAR) → Related to the missing value itself.

Choose a Handling Strategy

Remove → Drop rows/columns with too many missing values (e.g., >30%).

Impute → Fill missing values with estimates.

Flag → Add an indicator variable showing missingness (useful for ML models).

Apply the Chosen Method

Use appropriate imputation techniques depending on the data type and context.

Validate Results

Check distributions before and after imputation to ensure minimal distortion.

One Imputation Technique for Each Data Type
Numerical Data → Mean Imputation

Replace missing numerical values with the mean of the non-missing values.

Example: If exam scores are [85, NaN, 90, 80], replace NaN with mean = (85+90+80)/3 = 85.

Pros: Simple, quick, preserves dataset size.

Cons: Can reduce variance and bias results if data is not normally distributed.

Categorical Data → Mode Imputation

Replace missing categorical values with the most frequent category (mode).

Example: If favorite colors are [Red, NaN, Blue, Red], replace NaN with Red.

Pros: Easy to implement, preserves common patterns.

Cons: Can distort category proportions if too many are missing.


```

```python
#Question 8: Write a Python script to:
# Generate 1000 samples from a normal distribution.
# Introduce 50 random missing values.
# Fill missing values with the column mean.
# Plot a histogram before and after imputation.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Generate 1000 samples from a normal distribution
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # mean=50, std=10

# 2. Introduce 50 random missing values
missing_indices = np.random.choice(1000, 50, replace=False)
data[missing_indices] = np.nan

# Create DataFrame
df = pd.DataFrame({'Values': data})

# 3. Fill missing values with column mean
mean_value = df['Values'].mean()
df['Values_imputed'] = df['Values'].fillna(mean_value)

# 4. Plot histogram before and after imputation
plt.figure(figsize=(10, 5))

# Before imputation
plt.subplot(1, 2, 1)
plt.hist(df['Values'].dropna(), bins=30, edgecolor='black')
plt.title("Before Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

# After imputation
plt.subplot(1, 2, 2)
plt.hist(df['Values_imputed'], bins=30, edgecolor='black')
plt.title("After Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Ans-
1) Basic cleanup

Drop exact duplicates, fix types (e.g., amount numeric), standardize currency/units, trim string categories, unify category labels (e.g., “CC”, “Credit Card” → credit_card).

2) Missing data (e.g., missing ages)

Diagnose: quantify % missing and check if MCAR/MAR (e.g., is “age” missing more for certain payment methods?).

Impute (and flag):

Numerical (age):

Start with median imputation (robust to outliers).

Add a missingness indicator (age_missing=1/0).

If you have signal in other fields, consider KNNImputer or model-based imputation (e.g., train a small regressor on train fold).

Categorical (e.g., payment_method if missing):

Most frequent (mode) or fill with an explicit “Unknown” category (keeps information about missingness).

3) Outliers (e.g., transaction amount)

Detect: Boxplots/IQR, robust z-scores, or domain rules (negative/huge amounts).

Treat (pick 1–2 based on model choice):

Log transform: log1p(amount) to tame heavy tails.

Winsorize/cap using IQR or high percentiles (e.g., 1st/99th).

Robust scaling (Median/IQR) rather than StandardScaler.

Keep a cap flag if you winsorize.

Trees handle raw outliers better; linear/NN models usually benefit from log+robust scaling.

4) Target imbalance (fraud ≪ non-fraud)

Don’t “balance” your validation/test; keep real ratios.

Within training only (and inside each CV fold):

Class weights (e.g., class_weight='balanced') for LR/trees/XGBoost.

Resampling:

SMOTE/SMOTENC (if you have categorical features) to synthesize minority.

Or undersample majority (careful with information loss), or combine (SMOTE + TomekLinks).

Threshold tuning on validation to meet business costs (precision–recall trade-off).

Prefer PR-AUC, Recall@fixed Precision, F1 over accuracy.

5) Categorical encoding (e.g., payment method)

Low-cardinality: One-hot (handle unseen with handle_unknown='ignore').

High-cardinality (e.g., merchant_id): Target/mean encoding (with CV/fold-wise smoothing to avoid leakage) or CatBoost encoding.

Ordinal categories: use ordinal encoding only if true order exists.

Group rare levels into “Other” to stabilize.

6) Scaling & feature prep

Numerical:

If using linear/NN → RobustScaler (or StandardScaler after log).

For tree models, scaling is optional.

Date/time features: derive hour/day-of-week, is_weekend, time since last txn, rolling counts/amounts (e.g., #txns last 24h) — powerful for fraud.

Customer/merchant aggregates (computed only from past data in train folds): RFM, avg amount, velocity features.

7) Assemble a leak-proof pipeline

Use a single pipeline so all steps are learned from train only and reproduced at inference.


8) Cross-validation & checks

Stratified (Group) K-Fold for stable estimates with imbalance.

Monitor data drift between train/val (e.g., amount distribution).

After imputation/encoding, compare distributions to ensure no distortion.

9) Documentation & reproducibility

Fix random seeds, log pipeline versions/hyperparams, and save the entire fitted pipeline for deployment (so the same imputers/encoders run at inference).

TL;DR mapping to your issues

Missing ages → Median (or KNN) + missing flag.

Amount outliers → Log transform + robust scaling (or winsorize).

Imbalanced fraud → Class weights and/or SMOTENC (train-folds only) + threshold tuning; evaluate with PR-AUC.

Categoricals → One-hot for low-cardinality; target/CatBoost encoding for high-cardinality; handle unknowns.
